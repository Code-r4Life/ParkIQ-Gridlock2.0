```python
import pandas as pd

df = pd.read_csv(
    "/jan to may police violation_anonymized791b166 (2).csv",
    low_memory=False
)

# Drop useless columns
df = df.drop(
    columns=[
        "description",
        "closed_datetime",
        "action_taken_timestamp"
    ],
    errors="ignore"
)

# Convert datetime
df["created_datetime"] = pd.to_datetime(
    df["created_datetime"],
    errors="coerce"
)

# Remove invalid coordinates
df = df[
    (df["latitude"].between(-90,90))
    &
    (df["longitude"].between(-180,180))
]

# Remove rows with missing coordinates
df = df.dropna(
    subset=[
        "latitude",
        "longitude"
    ]
)

print(df.shape)

df.to_csv(
    "cleaned_violations.csv",
    index=False
)
```

```python
df["hour"] = df["created_datetime"].dt.hour

df["day_of_week"] = (
    df["created_datetime"]
    .dt.day_name()
)

df["month"] = (
    df["created_datetime"]
    .dt.month
)

df["weekend"] = (
    df["day_of_week"]
    .isin(["Saturday","Sunday"])
)

df["peak_hour"] = (
    df["hour"]
    .between(17,20)
)
```

```python
df.to_csv(
    "featured_violations.csv",
    index=False
)
```
